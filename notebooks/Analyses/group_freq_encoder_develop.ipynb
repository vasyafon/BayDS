```python
def group_freq_enc(df,groupby=[],cols=[], count=True, freq=False):
    slice_df = df[groupby+cols]
    if len(groupby) == 1:
        gcname=groupby[0]
    else:
        gcname = "("+'+'.join(groupby)+")"

    for igc,gc in enumerate(groupby):
        if igc == 0:
            slice_df['group_col'] = slice_df[gc].astype(str)
        else:
            slice_df['group_col'] += '_'+slice_df[gc].astype(str)
    if freq:
        count = slice_df['group_col'].map(slice_df['group_col'].value_counts(dropna=False)) 
    for col in cols:
        tempdf = (slice_df['group_col']+'_'+slice_df[col].astype(str))
        tempdf = tempdf.map(tempdf.value_counts(dropna=False))
        df[f'{col}_count_groupby_{gcname}'] = tempdf
        if freq:
            df[f'{col}_count_groupby_{gcname}_freq'] = tempdf / count
    return df

# smalldf = df[df['new_card_id'].isin([100,101])][['new_card_id','DT_M','id_16']]
# smalldf = group_freq_enc(smalldf,groupby=['new_card_id','DT_M'],cols=['id_16'],freq=True)
```
