## remove highly correlated and nan data

```python
df = p.data['data']
corr = df.corr()
```

```python
import seaborn as sns
sns.heatmap(corr, annot=True, fmt='.2f')
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(16,9))
plt.imshow(corr.values)
```

```python
threshold = 0.95
features_to_drop = set()
features = list(corr.columns)
for i in range(len(features)):
    for j in range(i+1,len(features)):
        c = corr[features[i]][features[j]]
        if c> threshold:
            
            if features[i] in features_to_drop or features[j] in features_to_drop:
                continue
            avg_corr_i = corr[features[i]].sum()
            avg_corr_j = corr[features[j]].sum()
            print (features[i],avg_corr_i ,features[j], avg_corr_j)
            if avg_corr_i>avg_corr_j:
                features_to_drop.add(features[i])
            else:
                features_to_drop.add(features[j])
```

```python
features_to_drop
```

```python
features = list(corr.columns)
for i in range(len(features)):
    nan_count_i = corr[features[i]].isna().sum()
    if nan_count_i > 500:
        print('Poor feature: ', features[i])
        features_to_drop.add(features[i])
```

```python
df_filtered = df.drop(list(features_to_drop),axis=1)
corr_filtered = df_filtered.corr()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(16,9))
plt.imshow(corr_filtered.values)
```
